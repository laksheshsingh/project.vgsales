```python
#Total Count of Games for each Platform

#PlayStation 2 and Nintendo DS have the most game titles in their platform. 
#In fact, they are almost equal in count, which makes it all the more interesting to see.
```

```python
plt.figure(figsize=(12,6), dpi=150)
sns.countplot(data=game,
              x="Platform",
              order = game["Platform"].value_counts().index,
              palette="Dark2")
plt.xticks(rotation=90)
plt.show()
```

```python
#Total Count of Games for each Genre

#Action seems to be the genre where most game title can be found,
#with Sports, Misc and Role-Playing following not too far behind.
#Puzzle appears to be the least popular among game developers.
```

```python
plt.figure(figsize=(12,6), dpi=150)
sns.countplot(data=game,
              x="Genre",
              order = game["Genre"].value_counts().index,
              palette="Dark2")
plt.xticks(rotation=90)
plt.show()
```

```python
#Explore correlation between features

#It is expected that all the sales columns are correlated to one another based on the heatmap below.
#There seems to be no significant correlation between the other columns.
```

```python
from sklearn.preprocessing import LabelEncoder

# Encode selected categorical columns to prepare for heatmap plotting
encoded_df = game.copy()
categ = ["Platform", "Genre", "Publisher"]
encoder = LabelEncoder()
encoded_df[categ] = encoded_df[categ].apply(encoder.fit_transform)

# Create correlation heatmap
plt.figure(figsize=(6,6), dpi=120)
sns.heatmap(encoded_df.corr(), 
            cmap="flare",
            annot=True)

plt.show()
```

```python
#Q1: What are the top 10 games making the most sales globally?
```

```python
# Using the `.nlargest()` method, we can easily extract the top n values of a dataframe
game.nlargest(10, "Global_Sales")
```

```python
#Q2: What are the top 5 gaming Genres that are making high sales?
```

```python
# We will use the `.groupby()`method to consolidate the indivisual genre, and the `.agg()` method to perform the sum function
sales_genre = game.groupby("Genre").agg({"Global_Sales": pd.Series.sum})
sales_genre.nlargest(5, "Global_Sales")
```

```python
#Q3: - Which game made the most sales per region and globally?
```

```python
print(f"NA: {game.nlargest(1, 'NA_Sales').values[0,1]}")
print(f"EU: {game.nlargest(1, 'EU_Sales').values[0,1]}")
print(f"JP: {game.nlargest(1, 'JP_Sales').values[0,1]}")
print(f"Other: {game.nlargest(1, 'Other_Sales').values[0,1]}")
print(f"Global: {game.nlargest(1, 'Global_Sales').values[0,1]}")
```

```python
#Q4:- Which Publishers made the most sales?
```

```python
sales_publisher = game.groupby("Publisher").agg({"Global_Sales": pd.Series.sum})
sales_publisher.nlargest(5, "Global_Sales")
```
